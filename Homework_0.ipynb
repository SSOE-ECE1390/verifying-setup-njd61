```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load and View Images

```python
### Load and View image
apollo_image = cv2.imread(os.path.relpath('./Data/Apollo_11_Launch.jpg'))
print(f"Image is {apollo_image.shape}")
print(f"Data Type is {apollo_image.dtype}")

# Fix Channels
apollo_image_rgb = apollo_image[:,:,::-1]
```

```python
## View Images
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

axes[0].imshow(apollo_image)
axes[0].set_title("Image in CV2 Format")

axes[1].imshow(apollo_image_rgb)
axes[1].set_title("Image in RGB Format")
```

### Crop the Image

```python
apollo_image_rgb_cropped = apollo_image_rgb[100:550, 400:800, :]
plt.imshow(apollo_image_rgb_cropped)
plt.title("Cropped Image")
```

```python

### Edit the brightness
mask = np.ones(apollo_image_rgb_cropped.shape, dtype=apollo_image_rgb_cropped.dtype) * 100
apollo_image_rgb_cropped_brighter = cv2.add(apollo_image_rgb_cropped, mask)

fig, axes = plt.subplots(1, 2, figsize=(8, 5))
axes[0].imshow(apollo_image_rgb_cropped)
axes[0].set_title("Before")

axes[1].imshow(apollo_image_rgb_cropped_brighter)
axes[1].set_title("After")
```

```python
### Rotate Brighter Images
apollo_image_rgb_cropped_brighter_rotated = cv2.rotate(apollo_image_rgb_cropped_brighter, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(apollo_image_rgb_cropped_brighter_rotated)
```

```python
### Save the image
cv2.imwrite(os.path.relpath('./Data/new_image.jpg'), apollo_image_rgb_cropped_brighter_rotated)
```
